```python
import pickle
import numpy as np
```

```python
with open('mfcc_LSTM.pkl', 'rb') as f:
    paradict = pickle.load(f)
x_train = paradict['x_train']
```

```python
def normalization(data):
    _range = np.max(data) - np.min(data)
    return (data - np.min(data)) / _range
```

```python
normalization(x_train)
```
